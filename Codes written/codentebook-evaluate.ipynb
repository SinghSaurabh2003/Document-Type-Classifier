```python
from torchvision.datasets import ImageFolder
from torchvision import transforms

IMAGE_SIZE = 224   # or import from config_v2

transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

test_dataset = ImageFolder(
    root="/kaggle/input/datasets/singhsaurabh03/testing/test",
    transform=transform
)

print(test_dataset.classes)
print(test_dataset.class_to_idx)
print("Total Images:", len(test_dataset))
```

```python
from pathlib import Path
from collections import Counter

train_path = Path("/kaggle/input/datasets/singhsaurabh03/training-half/Training_half")

labels = [
    "".join(filter(str.isalpha, p.stem))
    for p in train_path.glob("*.tif")
]

counter = Counter(labels)

print("=" * 50)
print("Training Dataset Distribution")
print("=" * 50)

for cls, count in sorted(counter.items()):
    print(f"{cls:<15} : {count}")
```
